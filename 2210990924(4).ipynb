# WHAT IS UNIVARIATE ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='green')
plt.title('Univariate Analysis: Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import matplotlib.pyplot as plt

# Group the data by the 'Survived' column and count the occurrences of each category
survived_count = data['Survived'].value_counts()

# Plot the counts
plt.figure(figsize=(8, 6))
plt.bar(survived_count.index, survived_count.values, color=['red', 'green'])
plt.xticks(survived_count.index, ['Died', 'Survived'])
plt.xlabel('Outcome')
plt.ylabel('Number of People')
plt.title('Number of People Survived and Died')
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
import pandas as pd

print("First Class:",len(data[data["Pclass"]==1]))
print("Second Class:",len(data[data["Pclass"]==2]))
print("Third Class:",len(data[data["Pclass"]==3]))

First_Class=len(data[data["Pclass"]==1])
Second_Class=len(data[data["Pclass"]==2])
Third_Class=len(data[data["Pclass"]==3])

import matplotlib.pyplot as plt

labels = ['First Class', 'Second Class', 'Third Class']
sizes = [ First_Class,Second_Class,Third_Class]
colors = ['yellow', 'pink', 'lightblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Passenger Distribution by Class')
plt.show()


```

# NUMBER  OF MALE AND FEMALE PASSANGERS

```python
class_count = data['Pclass'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(class_count, labels=class_count.index, autopct='%1.1f%%', colors=['blue', 'green', 'red'])
plt.title('Distribution of Passengers by Class')
plt.legend(labels=['1st Class', '2nd Class', '3rd Class'], loc='upper right')

plt.show()
```

# BIVARIANT ANALSIS

```python
data.columns
plt.figure(figsize=(8, 6))
plt.scatter(data['PassengerId'], data['Age'], color='blue', alpha=0.7)
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatter Plot of Variable 1 vs Variable 2')
plt.show()
```

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
survival_pivot = data.pivot_table(index='Sex', columns='Survived', aggfunc='size', fill_value=0)

# Plot the survival rates
survival_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Count by Sex')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xticks(rotation=0)
plt.show()
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
survival_pivot = data.pivot_table(index='Pclass', columns='Survived', aggfunc='size', fill_value=0)

# Plot the survival rates
survival_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Count by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```
